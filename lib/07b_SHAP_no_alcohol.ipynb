#### Calculate SHAP values and create plot with non alcohol consumers

```python
import shap 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline

import numpy as np
import pandas as pd
from pickle import load

import matplotlib.pylab as pl
import matplotlib.pyplot as plt

import os
os.chdir("../lib/")

shap.initjs()
```

```python
np.random.seed(0)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
x_train = pd.read_csv("../output/05/x_train_no_alc.csv", header = 0,index_col = 0)
x_test = pd.read_csv("../output/05/x_test_no_alc.csv", header = 0,index_col = 0)
y_train = pd.read_csv("../output/05/y_train_no_alc.csv", header = 0, index_col = 0)
y_test = pd.read_csv("../output/05/y_test_no_alc.csv", header = 0, index_col = 0)

x_train = x_train.sort_index(axis = 0)
x_test = x_test.sort_index(axis = 0)
y_train = y_train.sort_index(axis = 0)
y_test = y_test.sort_index(axis = 0)

mod = load(open("../model/rf_homa_no_alc.pkl", "rb"))
```

```python
print("Mod params: {}".format(mod))
```

```python
rf_mod = mod.best_estimator_.named_steps['model']
```

```python
feature_names = x_test.columns
```

```python
shap_values = shap.TreeExplainer(rf_mod).shap_values(x_test)
```

```python
shap.summary_plot(shap_values, x_test, 
                 feature_names = feature_names, max_display=20)
```
